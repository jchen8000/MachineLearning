# Support Vector Machines


***ABSTRACT***

This is a hands-on practice of Support Vector Machine (SVM) algorithm. The goal of Support Vector Machines (SVM) is to find a hyperplane that maximizes the minimum distance to the hyperplane. A hyperplane in 1-D is simply a point, in 2-D it’s a line, in 3-D it’s a plane, more than 3-D it's a hyperplane.

1. Understand the SVC algorithm by math.

2. Visualize the SVC with linear kernel function.

3. Visualize the SVC with non-linear, Gaussian kernel function.



```
%matplotlib inline
import numpy as np
from mpl_toolkits import mplot3d
import matplotlib.pyplot as plt
from scipy import stats
```

## 1. Algorithm

The goal of support vector machines is to find the line that maximizes the minimum distance to the line.

**Support Vector Machine**


$\qquad {\displaystyle\min_{\theta}\sum_{i=1}^m \left[ y^{(i)}cost_1(\theta^Tx^{(i)}) + (1-y^{(i)})cost_0(\theta^Tx^{(i)}) \right] }$ 

![alt text](https://raw.githubusercontent.com/jchen8000/MachineLearning/master/images/supportvectormachine.png)


**SVM Decision Boundary:**

$\qquad {\displaystyle\min_{\theta}\frac{1}{2}\sum_{j=1}^n\theta_j^2=\displaystyle\frac{1}{2}(\theta_1^2+\theta_2^2+...)}=\displaystyle\frac{1}{2}(\sqrt{\theta_1^2+\theta_2^2+...})^2 = \frac{1}{2}||\theta||^2$ 


$\qquad \textrm{ s.t. } \quad\, \theta^Tx\geqslant1 \quad \textrm{if} \quad y^{(i)}=1$

$\qquad  \quad \quad \quad \theta^Tx\leqslant1 \quad \textrm{if} \quad y^{(i)}=0$


**Gaussian Kernel Function:**

$\qquad  f_i=similarity(x, l^{(i)}) = e ^{-\displaystyle\frac{||x - l^{(i)} ||^2 }{2\sigma^2}} $

## 2. SVC with Linear Kernel

```
# importing scikit learn with make_blobs 
from sklearn.datasets.samples_generator import make_blobs 
  
# creating datasets X containing n_samples 
# Y containing two classes 
X, y = make_blobs(n_samples=500, centers=2, 
                  random_state=0, cluster_std=0.40) 


# plotting scatters  
#plt.figure(1, figsize=(8,6))
#plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter'); 
#plt.show()  
```

```
from sklearn.svm import SVC # "Support vector classifier"
svc = SVC(kernel='linear', C=1E10)
svc.fit(X, y)
```

```
def plot_svc_decision_function(model, ax=None):
    """Plot the decision function for a 2D SVC"""
    if ax is None:
        ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()
    
    # create grid to evaluate model
    x = np.linspace(xlim[0], xlim[1], 100)
    y = np.linspace(ylim[0], ylim[1], 100)
    Y, X = np.meshgrid(y, x)
    xy = np.vstack([X.ravel(), Y.ravel()]).T
    P = model.decision_function(xy).reshape(X.shape)
    
    # plot decision boundary and margins
    ax.contour(X, Y, P, colors='r',
               levels=[-1, 0, 1], alpha=0.5,
               linestyles=['--', '-', '--']
              )
    ax.set_xlim(xlim)
    ax.set_ylim(ylim)

def plot_support_vectors(model): 
    sv_color = 'goldenrod'
    for i in range(model.support_vectors_.shape[0]):
      plt.annotate('Support Vectors', xy=(model.support_vectors_[i]), color=sv_color, textcoords='axes fraction',xytext=(.8, .95),
                  arrowprops=dict(arrowstyle='->', connectionstyle='arc3,rad=-0.2', color=sv_color),
                 )
      plt.scatter(model.support_vectors_[:, 0],
                 model.support_vectors_[:, 1],
                 s=300, edgecolors=sv_color, facecolors='none');    
    
    
    
plt.figure(1, figsize=(8,6))
plt.scatter(X[:, 0], X[:, 1], c=y, s=50, cmap='winter')
plot_svc_decision_function(svc);
plot_support_vectors(svc);
plt.show() 
```

## 2. SVC with Gaussian Kernel

```
from scipy.stats import multivariate_normal

#Parameters to set
mu_x = 0
variance_x = 8

mu_y = 0
variance_y = 16

#Create grid and multivariate normal
x = np.linspace(-10,10,500)
y = np.linspace(-10,10,500)
X, Y = np.meshgrid(x,y)
pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X; pos[:, :, 1] = Y
rv = multivariate_normal([mu_x, mu_y], [[variance_x, 0], [0, variance_y]])

#Make a 3D plot
fig = plt.figure(figsize=(10, 8))
ax = fig.gca(projection='3d')
ax.grid(True)
ax.plot_surface(X, Y, rv.pdf(pos),cmap='viridis',linewidth=1)
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')
ax.set_title('Multivariate Gaussian function in 3D',fontsize=18)
plt.show()
```

```
from sklearn.datasets.samples_generator import make_circles
X_c, y_c = make_circles(300, factor=.2, noise=.1, random_state=0)

clf = SVC(kernel='rbf', gamma='auto', C=1E6).fit(X_c, y_c)

z = np.exp(-(X_c ** 2).sum(1))
```

```
fig = plt.figure(figsize=(20, 8))
#fig.subplots_adjust(left=0.0625, right=0.95, wspace=0.5)

ax = fig.add_subplot(1, 2, 1, projection='3d')
ax.scatter3D(X_c[:, 0], X_c[:, 1], z, c=y_c, s=50, cmap='viridis')
#ax.view_init(elev=elev, azim=azim)
ax.set_xlabel('$x_0$',fontsize=16)
ax.set_ylabel('$x_1$',fontsize=16)
ax.set_zlabel('z',fontsize=16)
ax.set_title('Transform input data to 3D with Gaussian function\nData can be separated by a 2D-hyperplane',fontsize=18)
ax.text(1.2, 1.2, 1.15, "$z=e^{-(x_0^2+x_1^2)}$", color='black',fontsize=16)

plt.subplot(1, 2, 2)
plt.scatter(X_c[:, 0], X_c[:, 1], c=y_c, s=50, cmap='viridis')
plot_svc_decision_function(clf);
plot_support_vectors(clf);

plt.show()
```
