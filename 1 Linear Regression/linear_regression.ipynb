## 2. Linear Regression

### 2.1 Hypothesis, Cost Function and Gradient Descent

**Algorithm of Linear Regression**

$\qquad \textrm{Assume } y|x;\theta\sim\mathcal{N}(\mu,\sigma^2)$

> **Hypothesis Function**: 

$\qquad\qquad  h_\theta(x) =  \theta_0 +  \theta_1x_1 + \theta_2x_2 + ... + \theta_nx_n \quad$,  Or it can be expressed in matrix: 

$\qquad\qquad h_\theta(x)  =  \theta ^ T x $

>  **Cost Function**:

$ \qquad\qquad J( \theta ) = \displaystyle\frac{\mathrm{1} }{2m}  \sum_{i=1}^{m} (h_ \theta (x^i) - y^i )^2 +    \frac{\lambda}{2m} \sum_{j=1}^{n} \theta_j^2   $

> **Gradient Descent**, which is the partial derivative of the cost function:

$ \qquad\qquad \displaystyle\frac{\partial J( \theta ) }{\partial \theta_0} = \frac{\mathrm{1} }{m}  \sum_{i=1}^{m} ( h_\theta(x^i) - y^i ) x^i_0  \qquad\qquad\qquad \textrm{ (where j = 0) }$

$ \qquad\qquad  \displaystyle\frac{\partial J( \theta ) }{\partial \theta_j} = \frac{\mathrm{1} }{m}  \sum_{i=1}^{m} ( h_\theta(x^i) - y^i ) x^i_j  + \frac{\lambda}{m} \sum_{j=1}^{n} \theta_j \qquad \textrm{(where j >= 1)}$ 

> **Normal Equation**:

$ \qquad\qquad \theta=(X^TX)^{-1}X^Ty$

**Normal Equation** is a quick way to calculate $\theta$ without the need for iterations, but it requires to calculate the Inverse of matrix, thus needs more computing cost and sometimes the matrix is not invertible, you can not resolve it.

Now we start with defining the functions as described above:

```
def prediction(X, theta):
    return X @ theta

def costFunction(X, y, theta, lmbda):
    temp = prediction(X, theta) - y
    return np.sum(np.power(temp, 2)) / (2*len(X)) + np.sum(theta[1:]**2) * lmbda / (2*len(X))
  
def gradientDescent(X, y, theta, alpha, lmbda, iterations):
    cost_history = [] # Cost history 
    theta_history = [] # Theta history
    m = len(y)
    for _ in range(iterations):
        temp = prediction(X, theta) - y
        temp = X.T @ temp
        theta = theta - (alpha/m) * temp
        theta = theta + theta * lmbda / m
        theta[0] = theta[0] - theta[0] * lmbda / m
        theta_history.append(list(theta[:,0]))
        cost_history.append(costFunction(X, y, theta, lmbda))
    return theta, cost_history, theta_history
```

### 2.2 Feature Scaling and Add Bias vector

```
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X = sc.fit_transform(X)

m = X.shape[0]
X = np.hstack((np.ones((m,1)), X))

```

### 2.2 Split the train and test data

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)

print(X.shape)
print(y.shape)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

### 2.3 Compute initial cost

```
m, n = X_train.shape
theta = np.zeros([n,1])
iterations = 500
alpha = 0.02
lmbda = 0.01


J = costFunction(X_train, y_train, theta, lmbda)
print(J)
```

### 2.4 Run Gradient Descent

```
theta, cost_history, theta_history = gradientDescent(X_train, y_train, theta, alpha, lmbda, iterations)
```

### 2.5 Final Cost ans Predictions

```
J = costFunction(X_train, y_train, theta, lmbda)

y_pred = prediction(X_test, theta)
R2_score = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print("Final Cost: ", J)
print("MSE: ", mse)
print("R2 Score:", R2_score)
```

### 2.6 Plot Results

```
text_font = {'family': 'serif',
        'color':  'darkred',
        'weight': 'normal',
        'size': 14,
        }

fig = plt.figure(1, figsize=(16,6))

plt.subplot(1, 2, 1)
plt.plot(range(len(cost_history)),cost_history,'b.')
plt.grid(True)

plt.text(300,260, "Final Cost =%f" % J, fontdict=text_font)
plt.title("Convergence of Cost", fontsize=20)
plt.xlabel("Iteration number", fontsize=16)
plt.ylabel("Cost", fontsize=16)
dummy = plt.xlim([-0.05*iterations,1.05*iterations])
dummy = plt.ylim([min(cost_history)-0.2*(max(cost_history)-min(cost_history)), max(cost_history)+0.2*(max(cost_history)-min(cost_history))])
    
  
plt.subplot(1, 2, 2)
sns.set(color_codes=True)
plt.text(10,43, "$R^2 = $%f" % R2_score, fontdict=text_font)
plt.text(10,41, "$MSE$ =%f" % mse, fontdict=text_font)
ax = sns.regplot(x=y_test, y=y_pred.reshape(y_pred.shape[0]), 
                 scatter_kws={"color": "blue"}, 
                 line_kws={"color": "red"})
ax.set_title( "Prices ($Y_i$) vs Predicted prices ($\hat{Y}_i$)", fontsize=20 )
ax.set_ylabel("Predicted prices: $\hat{Y}_i$", fontsize=16)
ax.set_xlabel("Prices: $Y_i$", fontsize=16)
ax.grid(True)
#plt.grid()
plt.show()
```
