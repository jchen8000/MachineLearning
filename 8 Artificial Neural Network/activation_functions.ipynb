```
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Sigmoid Function

```
def sigmoid(z):
  return 1/(1+np.exp(-z))

X = np.linspace(-5, 5, 100)
plt.plot([-5, 5], [0, 0], 'k-')
plt.plot([0, 0], [-0.05, 1.1], 'k-')
plt.plot(X, sigmoid(X),'b')
plt.xlabel('z')
plt.ylabel('$g(z)$')
plt.title('Sigmoid Function',fontsize=18)
plt.grid()
plt.text(-4, 0.9, r'$g(z)=\frac{1}{1+e^{-z}}$', fontsize=16)
plt.show()
```

## 2. Tanh Function

```
z = np.arange(-5, 5, .1)
t = np.tanh(z)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([-5, 5], [0, 0], 'k-')
ax.plot([0, 0], [-1, 1], 'k-')
ax.plot(z, t)
ax.set_ylim([-1.0, 1.0])
ax.set_xlim([-5,5])
ax.grid(True)
ax.set_xlabel('z')
ax.set_ylabel('$g(z)$')
ax.set_title('tanh function')
plt.text(-4, 0.6, r'$g(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$', fontsize=16)
plt.show()
```

## 3. ReLU Function 

```
# alternative activation function
def ReLU(x):
    return np.maximum(0.0, x)
# derivation of relu
def ReLU_derivation(x):
    if x <= 0:
        return 0
    else:
        return 1
      
      
X = np.linspace(-5, 5, 100)
plt.plot([-5, 5], [0, 0], 'k-')
plt.plot([0, 0], [-0.5, 5.2], 'k-')
plt.plot(X, ReLU(X),'b')
plt.xlabel('z')
plt.ylabel('$g(z)$')
plt.title('ReLU Function', fontsize=18)
plt.grid()
plt.text(-4.8, 4.5, r'$g(z)=max(0.0, z)$', fontsize=16)
plt.show()
```

## 4. Leaky ReLU Function

```
def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha*z, z)

z = np.linspace(-5, 5, 100)

plt.plot(z, leaky_relu(z, 0.05), "b-", linewidth=2)
plt.plot([-5, 5], [0, 0], 'k-')
plt.plot([0, 0], [-0.5, 4.2], 'k-')
plt.grid(True)
plt.xlabel('z')
plt.ylabel('$g(z)$')
props = dict(facecolor='red', shrink=0.1)
plt.annotate('Leak', xytext=(-3.5, 0.5), xy=(-5, -0.2), arrowprops=props, fontsize=14, ha="center", color='red')
plt.title("Leaky ReLU Function", fontsize=18)
plt.text(-4, 3.5, r'$g(z)=max(\epsilon z, z),$', fontsize=16)
plt.text(-4, 3., r'with $\epsilon\ll1$', fontsize=16)

plt.axis([-5, 5, -0.5, 4.2])

plt.show()
```
