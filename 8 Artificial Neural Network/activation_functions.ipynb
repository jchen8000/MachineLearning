```
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Sigmoid Function

Function:

$\qquad \displaystyle g(z)=\displaystyle\frac{1}{1+e^{-z}}$

Derivative:

$\qquad \displaystyle\frac{\mathrm{d} }{\mathrm{d} z}g(z) = g(z)(1-g(z)) $

```
def sigmoid(z):
  return 1/(1+np.exp(-z))

def sigmoidDerivative(z):    
    return sigmoid(z) * (1 - sigmoid(z) )

X = np.linspace(-5, 5, 100)

fig, ax = plt.subplots(1, 2, figsize=(16,5))
ax[0].arrow(-5, 0, 9.8, 0, head_width=0.04, head_length=0.5, fc='k', ec='k')
ax[0].arrow(0, -0.1, 0, 1.15, head_width=0.2, head_length=0.1, fc='k', ec='k')
ax[0].plot(X, sigmoid(X),'b')
ax[0].set_title('Sigmoid Function',fontsize=18)
ax[0].grid()
ax[0].set_ylim([-0.05, 1.2])
ax[0].text(0.1, 1.1, r'$g(z)=\frac{1}{1+e^{-z}}$', fontsize=16)
ax[0].text(4.95, 0.03, r'$z$', fontsize=16)

ax[1].arrow(-5, 0, 9.8, 0, head_width=0.04, head_length=0.5, fc='k', ec='k')
ax[1].arrow(0, -0.1, 0, 1.15, head_width=0.2, head_length=0.1, fc='k', ec='k')
ax[1].plot(X, sigmoidDerivative(X),'b')
ax[1].set_title('Sigmoid Derivative',fontsize=18)
ax[1].set_ylim([-0.05, 1.2])
ax[1].grid()
ax[1].text(0.1, 1.1, r"$g'(z) = g(z)(1-g(z))$", fontsize=16)
ax[1].text(4.95, 0.03, r'$z$', fontsize=16)

plt.show()
```

## 2. Tanh Function

Function:

$\qquad \displaystyle g(z)=\displaystyle\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$


Derivative:

$\qquad \displaystyle \frac{d}{dz}g(z) = 1 - g(z)^{2}$

```
z = np.arange(-5, 5, .1)
t = np.tanh(z)

dz = 1 - t**2

fig, ax = plt.subplots(1, 2, figsize=(16,5))
ax[0].arrow(-5, 0, 9.8, 0, head_width=0.06, head_length=0.5, fc='k', ec='k')
ax[0].arrow(0, -1.1, 0, 2.25, head_width=0.2, head_length=0.15, fc='k', ec='k')
ax[0].plot(z, t,'b')
ax[0].set_title('Tanh Function',fontsize=18)
ax[0].grid()
ax[0].set_ylim([-1.2, 1.4])
ax[0].text(0.1, 1.2, r'$g(z)=\frac{e^{z}-e^{-z}}{e^{z}+e^{-z}}$', fontsize=16)
ax[0].text(4.95, 0.03, r'$z$', fontsize=16)

ax[1].arrow(-5, 0, 9.8, 0, head_width=0.06, head_length=0.5, fc='k', ec='k')
ax[1].arrow(0, -1.1, 0, 2.25, head_width=0.2, head_length=0.15, fc='k', ec='k')
ax[1].plot(z, dz,'b')
ax[1].set_title('Tanh Derivative',fontsize=18)
ax[1].grid()
ax[1].set_ylim([-1.2, 1.4])
ax[1].text(0.1, 1.2, r"$g'(z)=1-g(z)^2$", fontsize=16)
ax[1].text(4.95, 0.03, r'$z$', fontsize=16)
plt.show()
```

## 3. ReLU Function 

Function:

$\qquad \begin{split}g(z) = \begin{Bmatrix} z & z \geqslant 0 \\
 0 & z < 0 \end{Bmatrix}\end{split}, \qquad \textrm{Or } \displaystyle g(z)=\textrm{max}(0,z) $
 
Derivative:

$\qquad\displaystyle\begin{split}\frac{d}{dz}g(z) = \begin{Bmatrix} 1 & z > 0 \\
0 & z<0 \end{Bmatrix}\end{split}$

```
# alternative activation function
def ReLU(x):
    return np.maximum(0.0, x)
    
# derivation of relu
def ReLU_derivation(x):
    return np.greater(x, 0).astype(int)


X = np.linspace(-4, 4, 100)

fig, ax = plt.subplots(1, 2, figsize=(16,5))
ax[0].arrow(0, -0.1, 0, 4.4, head_width=0.15, head_length=0.4, fc='k', ec='k')
ax[0].arrow(-4.2, 0, 8, 0, head_width=0.15, head_length=0.4, fc='k', ec='k')
ax[0].plot(X, ReLU(X),'b')
ax[0].set_title('ReLU Function',fontsize=18)
ax[0].grid()
ax[0].set_ylim([-0.2, 5])
ax[0].text(0.1, 4.5, r'$g(z)=max(0,z)$', fontsize=16)
ax[0].text(3.95, 0.1, r'$z$', fontsize=16)

ax[1].arrow(0, -0.1, 0, 4.4, head_width=0.15, head_length=0.4, fc='k', ec='k')
ax[1].arrow(-4.2, 0, 8, 0, head_width=0.15, head_length=0.4, fc='k', ec='k')
ax[1].plot(X, ReLU_derivation(X),'b')
ax[1].set_title('ReLU Derivative',fontsize=18)
ax[1].grid()
ax[1].set_ylim([-0.2, 5])
ax[1].text(0.1, 4.5, r"$g'(z)=1, z>0$", fontsize=16)
ax[1].text(0.9, 4.2, r"$=0, z<0$", fontsize=16)
ax[1].text(3.95, 0.1, r'$z$', fontsize=16)
plt.show()



```

## 4. Leaky ReLU Function

Function:

$\qquad\begin{split}g(z) = \begin{Bmatrix} z & z \geqslant 0 \\
 \epsilon z & z < 0 \end{Bmatrix}\end{split}, \qquad \textrm{Or } g(z)=max(\epsilon z, z),$
$\qquad \displaystyle \textrm{where } \epsilon\ll1$

Derivative:

$\qquad\displaystyle\begin{split}\frac{d}{dz}g(z)  = \begin{Bmatrix} 1 & z \geqslant 0 \\
\epsilon & z<0 \end{Bmatrix}\end{split}$

```
def leaky_relu(z, alpha=0.01):
    return np.maximum(alpha*z, z)

z = np.linspace(-5, 5, 100)

plt.plot(z, leaky_relu(z, 0.05), "b-", linewidth=2)
plt.plot([-5, 5], [0, 0], 'k-')
plt.plot([0, 0], [-0.5, 4.2], 'k-')
plt.grid(True)
plt.xlabel('z')
plt.ylabel('$g(z)$')
props = dict(facecolor='red', shrink=0.1)
plt.annotate('Leak', xytext=(-3.5, 0.5), xy=(-5, -0.2), arrowprops=props, fontsize=14, ha="center", color='red')
plt.title("Leaky ReLU Function", fontsize=18)
plt.text(-4, 3.5, r'$g(z)=max(\epsilon z, z),$', fontsize=12)
plt.text(-4, 3., r'with $\epsilon\ll1$', fontsize=12)

plt.axis([-5, 5, -0.5, 4.2])

plt.show()
```

## 5. Exponential Linear Unit (ELU) Function

Function:

$\qquad g(z) = \begin{split}\begin{Bmatrix} z & z \geqslant 0 \\ \alpha ( e^z - 1) & z < 0 \end{Bmatrix}\end{split}$

Derivative:

$\qquad\displaystyle\frac{d}{dz}g(z) = \begin{Bmatrix} 1 & z \geqslant 0 \\ \alpha e^z & z<0\end{Bmatrix}$


```
def elu(z, alpha=1):
    return np.where(z < 0, alpha * (np.exp(z) - 1), z)

z = np.linspace(-5, 5, 200)

plt.plot([-5, 5], [0, 0], 'k-')
plt.plot([-5, 5], [-0.5, -0.5], 'k--')
plt.plot([0, 0], [-2.2, 3.2], 'k-')
plt.plot(z, elu(z, alpha=0.5), "b-", linewidth=2)
plt.grid(True)
plt.title(r"ELU Function ($\alpha=0.5$)", fontsize=16)
#plt.title(, fontsize=16)
plt.text(-4.5, 2.5, r"$g(z)=z, \quad z \geqslant 0$", fontsize=12)
plt.text(-4.5, 2, r"$g(z)=\alpha(e^z-1) , \quad z < 0$", fontsize=12)

plt.axis([-5, 5, -2.2, 3.2])

plt.show()
```
